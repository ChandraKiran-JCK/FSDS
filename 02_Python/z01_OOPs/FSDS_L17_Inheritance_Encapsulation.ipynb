```python
class PFileOperations:

    def read(self):
        """
        This function reads data from a respective file.
        """
        global f1
        try:
            f1 = open(
                file=input("Enter file name: "),
                mode="r",
            )
            data = f1.read()
            return data
        except FileNotFoundError as e:
            print(e)
        finally:
            f1.close()

    def write(self):
        """
        This function writes data from a respective file.
        """
        global f2
        try:
            f2 = open(
                file=input("Enter file name: "),
                mode="w",
            )
            return f2.write(input("Enter the data you want to write: "))
        except FileNotFoundError as e:
            print(e)
        finally:
            f2.flush()
            f2.close()

class CFileOperations(PFileOperations):
    pass
```

```python
# access modifiers
# public, private & protected concept.

# the above concept can be achieved using notations.
# variableName/methodName -> public
# _variableName/_methodName -> protected
# __variableName/__methodName -> private

# public -> Accessible from anywhere inside or outside the class.
# protected -> Protected members are suggested for use only within the defined class and its child-class.
# private -> Private members are suggested for use only within the defined class.
```

```python
class MyClass:
    def public_method(self):
        return "This is public"

obj = MyClass()
print(obj.public_method()) # Works
```

```python
class MyClass:
    def _protected_method(self):
        return "This is protected"

class SubClass(MyClass):
    def access_protected(self):
        print("This is child class using protected method")
        return self._protected_method() # Works within subclass

obj1 = SubClass()
print(obj1.access_protected())

obj = MyClass()
print(obj._protected_method()) # Works, but discouraged by convention
```

```python
class MyClass:
    def __private_method(self):
        return "This is private"

    def access_private_internal(self):
        print("This is public method accessing private method")
        return self.__private_method() # Works inside the class

obj = MyClass()
print(obj.access_private_internal()) # Works

# print(obj.__private_method()) # This would raise an AttributeError

# You can technically access it via the mangled name:
print(obj._MyClass__private_method()) # works
```

```python
"""
Encapsulation
1. code security
2. bind the data as private members
3. create setter(data) and getter() public methods in-order to access the private members.
"""
```

```python
class Account:
    def __init__(self):
        self.__balance:float = 0
        self.a = 10
        self._b = 20
        self.__c = 30

    def setBalance(self , value):
        self.__balance = value

    def getBalance(self):
        return self.__balance

a1 = Account()
print(a1.getBalance())

# a1.__balance = 10 -> error
# a1._Account__balance = 10 -> works but not suggested

a1.setBalance(100)
print(a1.getBalance())

print(a1.a)
print(a1._b)
# print(a1.__c) -> error
print(a1._Account__c)

# Note :
# public , private , protected all these data members can be accessed but its not
# suggested to access them.
```

```python
"""
multiple inheritance
"""

class abc:
    def __init__(self , a1 , b1 , c1):
        self.a = a1
        self.b = b1
        self.c = c1

    def test(self):
        print("this is test method from abc class" , self.a, self.b, self.c)

class xyz:
    def __init__(self , x1 , y1 , z1):
        self.x = x1
        self.y = y1
        self.z = z1

    def test1(self):
        print("this is test method from xyz class" , self.x, self.y, self.z)


obj1 = abc(1,2,3)
obj1.test()

obj2 = xyz(7,8,9)
obj2.test1()

class child1(abc , xyz):
    pass

class child2(xyz , abc):
    pass

# c1 = child1()
# TypeError: abc.__init__() missing 3 required positional arguments: 'a1', 'b1', and 'c1'

# c2 = child2()
# TypeError: xyz.__init__() missing 3 required positional arguments: 'x1', 'y1', and 'z1'

"""
here child1 , child2 is inheriting both abc , xyz class but while creating object iam getting
2 different error bcz internally whenever we are inheriting multiple classes then, methods of both the
classes gets inherited but when comes to variables, it inherits only the first class mentioned.
shown as below.

class child1(abc , xyz):
    # __init__(self , a1 , b1 , c1) inherited
    # test()
    # test1()
    pass

class child2(xyz , abc):
    # __init__(self , x1 , y1 , z1) inherited
    # test()
    # test1()
    pass

now , all methods are getting inherited so no problem ,
but variables of only 1st class are getting inherited how to resolve this
so that variables of all the classes (a1,b1,c1,x1,y1,z1) gets inherited.
"""
```

```python
class child3(abc , xyz):

    # test()
    # test1()
    # init(a,b,c) -> default one so override this

    def __init__(self , a1 , b1 , c1 , x1 , y1 , z1):
        abc.__init__(self , a1 , b1 , c1)
        xyz.__init__(self , x1 , y1 , z1)
```

```python
c3 = child3(1,2,3,7,8,9)
c3.test()
c3.test1()
```

```python
class child4(abc , xyz):
    # over-ride the init(self)

    def __init__(self , *args):
        abc.__init__(self , args[0] , args[1] , args[2])
        xyz.__init__(self , args[3] , args[4] , args[5])
```

```python
c4 = child4(1,2,3,7,8,9,10,11)
c4.test()
c4.test1()
```
