```python
class insta :
    def share_stories(self):
        print("sharing my insta stories")

class facebook :
    def share_stories(self):
        print("sharing my facebook stories")
```

```python
def share(obj):
    obj.share_stories()
```

```python
i = insta()
f = facebook()
```

```python
share(i)
# """
# i -> insta class object
# share(insta_obj)
# insta_obj.share_stories()
# """
```

```python
share(f)
```

```python
"""
share(obj) -> behaviour is getting changed based on the object passed to it.
"""
```

```python
"""
Polymorphism with inheritance

1. compile-time polymorphism = method overloading.
    |=> same class , same name but different parameters.
2. run-time polymorphism = method overriding.
    |=> inheritance , same name & same parameters but in different class.
"""
```

```python
class social_media:
    def share_stories(self):
        print("share story to social media")
    def upload_pic(self):
        print("upload pic to social media")
```

```python
class insta(social_media):
    # share_stories & upload_pic

    def share_stories(self):
        print("share story to insta")

class facebook(social_media):
    # share_stories & upload_pic

    def share_stories(self):
        print("share story to facebook")
```

```python
i = insta()
f = facebook()
s = social_media()
```

```python
i.share_stories()
```

```python
f.share_stories()
```

```python
s.share_stories()
```

```python
"""
same function share_stories() but acting different for different objects.
"""
```

```python
"""
Custom Exceptions:
1. inherit Exception class
2. over-ride init() and make customization
"""
```

```python
class my_exception(Exception):
    def __init__(self , err_msg):
        self.err_msg = err_msg
```

```python
try :
    raise my_exception("This is my_exception got raised")
except my_exception as err:
    print(err.err_msg)
```

```python
class my_exception1(Exception):
    def __init__(self , err_msg , err_msg1):
        self.err_msg = err_msg
        self.err_val = err_msg1
```

```python
try :
    raise my_exception1("this is my custom exception2" , 10)
except my_exception1 as err:
    print(err.err_msg , err.err_val)
```

```python
"""
static members:
object data -> inside init(self)
    |=> when object is created then only memory allocation is done.
    |=> can be accessed only through object.

class data -> inside class outside init(self) ==> static data (common data shared among objects)
    |=> when class gets loaded then memory allocation is done.
    |=> as this is a class data we can access through object & class_name

static methods:
    |=> use decorator for method as '@staticmethod' and 'remove self' from method definition.
    |=> Example :
            a. normal method => object function
            def add(self , a , b):
                return a + b

            b. static method => class function
            @staticmethod
            def add(a , b):
                return a + b

"""
```

```python
"""
Abstraction -> template/blue-print of a class.
1. to implement abstraction we need to import abc module. abc(abstract base class)
2. from this abc module also import ABC class and abstractmethod.
3. make the class Inherit from ABC to make it an abstract class.
4. for each method inside mention abstract method.
"""
```

```python
import abc
from abc import ABC , abstractmethod

class data_project(ABC):
    # to make this class as abstract class/template make this class as child to ABC class.
    # inside abstract class make all methods as abstract methods by using @abstractmethod decorator for the methods.

    @abstractmethod
    def method1(self):
        pass

    @abstractmethod
    def method2(self):
        pass

    @abstractmethod
    def method3(self):
        pass

```

```python
# dev1 to work for method 1,2 and dev2 to work for method 3

class dev1(data_project):
    def method1(self):
        print("method 1 implementation by dev 1")
    def method2(self):
        print("method 2 implementation by dev 1")
    def method3(self):
        pass

class dev2(data_project):
    def method1(self):
        pass
    def method2(self):
        pass

    def method3(self):
        print("method 3 implementation by dev 2")
```

```python
a = data_project() # for a template/abstract class we can't create an object.
```

```python
b = dev1()
b.method1()
b.method2()
b.method3()
```

```python
c = dev2()
c.method1()
c.method2()
c.method3()
```
