```python
# OOPS -> object-oriented programming system.
# class | object
# class -> representative of objects | blueprint for an object.
# object -> real entity.
#    |-> property => variables.
#    |-> tasks    => functions.

'''
example:
Animal -> class (representative of animals i.e, tiger , lion , .....)
Tiger  -> object (real entity)
    properties -> 4legs , 2eyes , eye color , .....
    tasks      -> run() , eat() , sleep() .......
'''
```

```python
# constructor -> object initialization.
# __init__() -> in-built function acts as a constructor.
```

```python
# every function inside a class will have self as the 1st parameter inorder to point to the object.
# any function starts and ends with __ then it means it's an in-built function.
```

```python
class Car:
    # properties(variables)
    brand_name: str = ""
    fuel_type: str = ""
    body_count: int = -1
    wheels: int = -1
    color: str = ""

    # tasks
    def start(self):
        print("start the car")

    def stop(self):
        print("stop the car")

    def car_desc(self):
        print(
            "Brand Name :", self.brand_name, " ",
            "Fuel Type :", self.fuel_type, " ",
            "Car Capacity :", self.body_count, " ",
            "Wheels :", self.wheels, " ",
            "Car color :", self.color, " "
        )
```

```python
car1 = Car()
car1.start()
car1.stop()
car1.car_desc()
```

```python
car2 = Car()
# properties
car2.brand_name = "Tata"
car2.fuel_type = "Petrol"
car2.body_count = 7
car2.wheels = 4
car2.color = "Black"

# tasks
car2.start()
car2.stop()
car2.car_desc()
```

```python
car3 = Car()
# properties
car3.brand_name = "Audi"
car3.fuel_type = "petrol"
car3.body_count = 4
car3.wheels = 4
car3.color = "Grey"

# tasks
car3.start()
car3.stop()
car3.car_desc()
```

```python
"""
- here in above case we are 1st creating an object then later we are assigning the values to the object.
- we can use 'constructor' for assigning values to the object at the time of object creation.
"""
```

```python
class Student:
    #properties
    std_id: int = -1
    std_name: str = ""
    std_marks: int = ""
    std_address: str = ""

    # using constructor - to assign the values to the object(self-object).
    def __init__(self , stdId:int , stdName:str , stdMarks:int , stdAdd:str):
        self.std_id = stdId
        self.std_name = stdName
        self.std_marks = stdMarks
        self.std_address = stdAdd

    # tasks
    def read(self):
        print(self.std_name , " is reading")

    def sleep(self):
        print(self.std_name , " is sleeping")

    def std_description(self):
        std_details = [self.std_id , self.std_name , self.std_marks]
        return std_details
```

```python
s1 = Student()
```

```python
s1 = Student(1 , "Chandra" , 85 , "Vzm")
s1.read()
s1.sleep()
print(s1.std_description())

print("-------------------------------")

s2 = Student(2 , "Kiran" , 87 , "Vizag")
s2.read()
s2.sleep()
print(s2.std_description())

print("------------------------------")

s3 = Student(3 , "Jami" , 92 , "AP")
print(s3.std_description())
```

```python
class Animal:
    legs:int = -1
    eys:int = -1
    category:str = ""

    def __init__(self , leg , eye , cate):
        self.legs = leg
        self.eys = eye
        self.category = cate

    def desc(self):
        print(self.legs, self.eys, self.category)

# properties - variables/data
#  tasks - data processing/functions
```

```python
a1 :Animal = Animal()
a1.desc()
```

```python
a2= Animal()
a2.legs = 4
a2.eys = 2
a2.category="lion"
a2.desc()
```

```python
a3 = Animal(4 , 20 , "sdfghj")
a3.desc()
```
