```python
# WAF to take a list as input and return a list with only integers
l1 = [3.6 , 4 , 5+6j , 6 , 7 , 7 , "chandra" , 6 ,3 , 4 , 2.5 , 6 , [1,2,3,4,56,"hbjkjn"]]
# o/p = [4 , 6 , 7 , 7]
```

```python
def filter_integers(l):
    int_l = []
    if(type(l) == list):
        for i in l :
            if(type(i) == int):
                int_l.append(i)

    return int_l       
```

```python
l1
```

```python
filter_integers(l1)
```

```python
filter_integers("chanfta")
```

```python
type(filter_integers([1,2,3]))
```

```python
filter_integers([1,2.5,3])
```

```python
filter_integers()
```

```python
filter_integers([1,2,3] , "chabdra")
```

```python
# WAF to take dict as input and return keys as list 1:00
```

```python
d1 = {
    '4' : [1,2,3,4] ,
    "chandra" : 3+4j ,
    "123" : (1,2,3,4) , 
    "cfhgvjb" : "etryv"
}
```

```python
def dict_keys(d : dict):
    if(type(d) == dict):
        d_keys = d.keys()
        return d_keys
    else:
        return 'invalid input'
```

```python
dict_keys(d1)
```

```python
dict_keys("cvhbj")
```

```python
# WAF to take 2 inputs , if both the inputs are lists then merge them and return final list else return invalid inputs
def merge(a , b):
    if(type(a) != list or type(b) != list):
        return 'invalid input'
    else:
       return a + b
```

```python
merge([1,2,3] , [1,2,3])
```

```python
merge([1,2] , [3,4,2,1,0])
```

```python
merge([1,2] , "cnara")
```

```python
merge([1,2])
```

```python
def merge1(a , b):
    if(type(a) != list or type(b) != list):
        return 'invalid input'
    else:
       a.extend(b) # a gets extended
       return a
```

```python
merge1([1,2,3] , [1,2,3])
```

```python
merge1([1,2] , [3,4,2,1,0])
```

```python
# WAF to accept the size and print a star pattern based on the size given
def right_angeled_star(n):
    for i in range(1 , n+1 , 1):
        for j in range(1 , n+1 , 1):
            if(i >= j):
                # print star
                print('*' , end=" ")
        print()
```

```python
right_angeled_star(5)
```

```python
right_angeled_star(7)
```

```python
# Docstring (description of the function) -> inside the function use """ doc string """
```

```python
def right_angeled_star1(n):
    """
    This is function that takes a numeric value as input and based on the value a right angeled triangle is formed 
    """
    for i in range(1 , n+1 , 1):
        for j in range(1 , n+1 , 1):
            if(i >= j):
                # print star
                print('*' , end=" ")
        print()
```

```python
right_angeled_star1(5)
```
