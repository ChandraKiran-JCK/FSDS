```python
s = "chandra"
for i in s:
    print(i , end=" ")
    
# string is a iterable object(index-based) but here we are not using indexex to access , so internally for-loop converts 
# iterable string object into iterator string object and uses pointer based access.
```

```python
# list element insertion - append(element)
# tuple element insertion - not possible
# set element insertion - add(element)
# dict element insertion - assignment => dict[element] = value
```
