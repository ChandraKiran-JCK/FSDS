```python
# numeric list 
l = [1 , 4 , 5.6 , 4+7j]
# add 2 for each element
for i in l :
    uv = i+2 
    print(i , "update value is" , uv)
```

```python
# instead of individual element return a list 
l = [1 , 4 , 5.6 , 4+7j]
ul = []
for i in l :
    ul.append(i+2)

print(ul)
```

```python
l = [2 , 45 , 78 , 12 , "sudh" , 6+7j , [56,67,78] , 23 , 56]
# filter only integers from this list
for i in l:
    if(type(i) == int):
        print(i)
        
```

```python
# for-else 
# end statement after for loop
l = [2 , 45 , 78 , 12 , "sudh" , 6+7j , [56,67,78] , 23 , 56]
# filter only integers from this list
for i in l:
    if(type(i) == int):
        print(i)
else:
    print("for loop completed")
```

```python
# nested loop
l = [2 , 45 , 78 , 12 , "sudh" , 6+7j , [56,67,78,"chc",7.3] , 23 , 56]
# filter only integers from this list
for i in l:
    if(type(i) == int):
        print(i)
    elif(type(i) == list):
        for j in i:
            if(type(j) == int):
                print(j)
else:
    print("loop completed")        
```

```python
# practice
l = [2 , 45 , 78 ,"rtyui", 12 , 12 , 12 , 12 , "chandra" , 6+7j , [56 , 67,78,"ghjk"]]
```

```python
# q2. Try to extract all the list of char if element is a string
cl = []
for i in l :
    if(type(i) == str):
        for j in i :
            cl.append(j)

print(cl)
```

```python
# q3. Try to return a list after doing a square of all the int elements
sl = []
for i in l:
    if(type(i) == int):
        sl.append(i*i)

print(sl)
```

```python
# q1. Try to print index of all the element
# using range()
for i in range(0,len(l)):
    print("element " , l[i] , " index is " , i)
```

```python
l
```

```python
# note : if duplicate elements exists then index() returns wrong indexes
for i in l:
    print("element " , i , " index is " , l.index(i))
```

```python
# enumerate(collection) -> returns tuples of each index,value
for i in enumerate(l):
    print(i)
```

```python
t = tuple(l)
print(t)
for i in enumerate(t):
    print(i)
```

```python
# for loop with collection -> for i in collection 
# for loop with else       -> for loop completed completly then only else block
# for loop with range()    -> for i in range(SI , EI , Step)
# for loop with enumerate(collectiion) -> for i in enumerate(collectiion)
```
