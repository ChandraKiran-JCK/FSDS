```python
# To handle these type of problem we use Exception handling.
# 1. write the unsafe code in try block.
# 2. if exception/error raised then handle that error in except block.
```

```python
try:
    a = 10
    b = int(input("Enter a value"))
    print(a/b)
    print("done smooth")
except:
    print("invalid input from user")
    print("done")

print("my rest code")
    
```

```python
# Enter a value 5
# 2.0
# done smooth
# my rest code

# Enter a value 0
# invalid input from user
# done
# my rest code

# there is no exception raised as we have handled the exception using try-except blocks.
```

```python
# Runtime exception so handle using try-except block
try:
    for i in range(len(l)+1):
        print(l[i])
except :
    print("Their is some issue")

print("rest of the code")
```

```python
# in-order to know the exception occurred in try block, we can know that in except block.
try:
    for i in range(len(l)+1):
        print(l[i])
except Exception as e:
    print("Their is some issue and the issue is : " , e)

print("rest of the code")
```
