```python
"""
the above code failed ,bcz we have handled only ValueError and we haven't handled any other error's.
so to handle all type of error's , we need to maintain a handler with "Exception" class (as this is the parent of
all the exceptions that occurs) after all the handlers.
"""
```

```python
try:
    # dict operation
    d1 = {
        "key1" : "chandra" , 
        "key2" : [1,2,3,4,5] ,
        "key3" : (4,5,6,7)
    }
    d1["key4"] = int(input("Enter the value for key 4 : "))
    print(d1)

    # file operation
    f = open("text2.txt" , "r")
    f.write("dfghj fghjk")
    
except ValueError as e :
    print("Value Exception raised as : " , e)
except Exception as e :
    print("Exception raised as : " , e)

print("Rest of my code")
```

```python
# try-except-else block
# else -> try block executed successfully then only else block gets executed.
```

```python
try:
    d1 = {
        "key1" : "chandra" , 
        "key2" : [1,2,3,4,5] ,
        "key3" : (4,5,6,7)
    }
    d1["key4"] = int(input("Enter the value for key 4 : "))
    print(d1)
    
except Exception as e :
    print("Exception raised as : " , e)
else:
    print("try block executed succesfully")

print("Rest of my code")
```

```python
# execution flow 
# try(success) - except(skip) - else(execute) - rest code
# try(fail) - except(execute) - else(skip) - rest code 
```

```python
# finally block
```

```python
# irrespective of the flow finally-block gets executed.
```

```python
try:
    print("try block start")
    a = int(input("Enter a value : ")) # ValueError possible
    print(10/a) # divide with zero error possible
    print("try block end")

except ValueError as e:
    print("value error raised as : " , e)
except ZeroDivisionError as e:
    print("division with zero error raised as : " , e)
except Exception as e:
    print("Exception raised as : " , e)

else:
    print("try-block executed succesfully so else block")
finally :
    print("finally block")

print("rest of the code")
```

```python
# try block start
# Enter a value :  2
# 5.0
# try block end
# try-block executed succesfully so else block
# finally block
# rest of the code

# try block start
# Enter a value :  sdfg
# value error raised as :  invalid literal for int() with base 10: 'sdfg'
# finally block
# rest of the code

# try block start
# Enter a value :  0
# division with zero error raised as :  division by zero
# finally block
# rest of the code
```

```python
import logging

logging.basicConfig(
    filename="test3.log" , 
    format='{asctime} - {levelname} - {message}',
    level=logging.INFO,
    style='{', # <-- Uses brace formatting
)
```

```python
# log messages with function
def divbyzero(a,b):
    logging.info("divbyzero function start with value %s and %s " , a , b)
    try:
        res = a/b
        logging.info("function exected sucessfully with result as %s" , res)
    except Exception as e:
        logging.error("error occured " + str(e))
```

```python
logging.info("function call start")
divbyzero(4,5)
divbyzero(4,0)
logging.info("function call end")
```
