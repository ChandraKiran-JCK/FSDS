```python
# for loop with collection -> for i in collection 
# for loop with else       -> for loop completed completly then only else block
# for loop with range()    -> for i in range(SI , EI , Step)
# for loop with enumerate(collectiion) -> for i in enumerate(collection)
```

```python
l = [1, 2, 3, 4, 5]
for i in l :
    print(i)
```

```python
for i in l :
    print(i)
else:
    print('for loop completed')
```

```python
for i in l :
    if(i == 2):
        break
    print(i)
else :
    print('for loop completed')
```

```python
# for loop not completed completly
```

```python
for i in l:
    if(i == 7):
        break 
    print(i)

else:
    print('for loop completed')
```

```python
# -> while loop
#  while(condition):
#      true block
```

```python
a = 1
while(a < 6):
    print(a)
    a = a+1
```

```python
# while loop with break
a = 1
while(a < 6):
    if(a == 4):
        break
    print(a)
    a = a + 1
```

```python
# break -> break the loop in b/w
# continue -> skip the statements below for that iteration
```

```python
a = 1
while(a < 6):
    if(a == 3):
        a = a + 1
        continue
    print(a)
    a = a + 1
```

```python
# while loop without body
while(a < 5):
```

```python
# while loop wihout body then use pass statement
while(a < 5):
    pass
```

```python
# for loop without body 
for i in l:
```

```python
# for loop without body then use pass statement 
for i in l:
    pass
```

```python
# *
# **
# ***
# ****
# *****

for i in range(1,6):
    for j in range(1,6):
        if(i >= j):
            print("*" , end="")
    print()
            
```

```python
# range() -> generates data
# range(start , end , step)
# default start = 0 , default step = 1
```

```python
l = list(range(10)) # generates data form 0 till 9 => [0 , 10)
print(l)
```

```python
l = list(range(2,10))
print(l)
```

```python
l = list(range(1,10,2)) # [1 , 10) step=2
print(l)
```

```python
l = list(range(3,50,2))
print(l)
```

```python
d = {
    'a' : "dfgh" ,
    'b' : "erty" , 
    'c' : [1,2,3,4,5] ,
    'd' : (4,5,6,7) ,
    'e' : "chandra"
}

key = d.keys()
for i in key: # 'a' , 'b' , 'c', 'd' , 'e'
    print(i , '->' , d.get(i))
```

```python
d.items()
```

```python
for i in d.items(): # ('a', 'dfgh'), ('b', 'erty'), ('c', [1, 2, 3, 4, 5]), ('d', (4, 5, 6, 7)), ('e', 'chandra')
    print(i[0] , '->' , i[1])
```
