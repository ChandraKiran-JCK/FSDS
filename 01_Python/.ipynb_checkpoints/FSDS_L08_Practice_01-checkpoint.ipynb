```python
# 1. count the letters & spaces in the string 
s = 'This is basic python class'
lc = 0 
sc = 0
tc = 0
for i in s :
    tc = tc + 1
    if(i == ' '):
        sc = sc + 1
    else:
        lc = lc + 1
else:
    print("parsing completed")

print('letters count :' , lc)
print('spaces count :' , sc)
print('total length :' , tc)
```

```python
# 2. print the string in reverse order 
for i in range(len(s)-1 , -1 , -1): # i : indexes
    print(s[i] , end="")
```

```python
# using while loop
i = len(s)-1 # last index
while(i >= 0):
    print(s[i] , end="")
    i = i - 1
```

```python
# 3. find the vowels in the string
s = "ineuron"
v = "AaEeIiOoUu"
for i in s:
    if(i in v):
        print(i , "-> Vowel")
    else:
        print(i , "-> not Vowel")
else:
    print("parsing completed")
```

```python
# 4. String is palindrome or not
s = input("Enter a string")
str_rev = s[::-1]
if(s == str_rev):
    print(s, " is palindrome")
else:
    print(s, " is not palindrome")
```

```python
# using loop
s = input("Enter a string") # eye
str_rev = ""

for i in range(len(s)-1 , -1 , -1): 
    str_rev = str_rev + s[i]

print("reverse ->", str_rev)
    
if(s == str_rev):
    print(s, " is palindrome")
else:
    print(s, " is not palindrome")
```

```python
d = {
    'india' : "IN" ,
    'canada' : "CA" ,
    'china' : "CH" ,
    'united states' : "USA"
}
# dict with country and country code , find india is available in this dict or not 
countries = d.keys()
print("India".lower() in countries)
print("india".lower() in countries)
print("Indian".lower() in countries)
```

```python
# filter this dict into 2 lists based on key length <=5 , >5
keys = d.keys()
l1 = [] # keys length <=5
l2 = [] # keys length >5

for i in keys: 
    if(len(i) <=5):
        l1.append(i)
    else:
        l2.append(i)

print(l1)
print(l2)
```

```python
# max value for each key in a nested dict

d1 = {
    'ineuron':{
        "a" : 14 ,
        "b" : 10 ,
        "c" : 4
    } ,
    'course' : {
        "d" : 45 ,
        "e" : 34 ,
        "f" : 1
    }
}

for i in d1.values(): # [{'a': 14, 'b': 10, 'c': 4}, {'d': 45, 'e': 34, 'f': 1}]
    print(type(i) , '->' ,i)
    inner_values = i.values()
    print(inner_values)
    print(max(inner_values))
```

```python
# max integer among all the integers in a dict
d2 = {
    'ineuron':{
        "a" : 14 ,
        "b" : 10 ,
        "c" : 4
    } ,
    'course' : {
        "d" : 45 ,
        "e" : 34 ,
        "f" : 1
    } , 
    "g" : 354+7j ,
    "h" : [45, 6, 7, 8, 9, 3 , 6+7j] ,
    "i" : (45, 34, 2 , "chandra") ,
    "k" : "chandra" , 
    "l" : {3,45,45,2,3,45,3}
}

l = []

for i in d2.values():
    print(type(i) , '->' ,i)
    if(type(i) == int):
        l.append(i)
    elif(type(i) == list or type(i) == tuple or type(i) == set):
        data = i
        for inner_value in data :
            if(type(inner_value) == int):
                l.append(inner_value)
    elif(type(i) == dict):
        values = i.values()
        for inner_value in values :
            if(type(inner_value) == int):
                l.append(inner_value)
else:
    print('dict traversal completed')

print(l)
print(max(l))
```
