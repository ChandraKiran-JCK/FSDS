```python
# so we have 2 solutions
# 1. place the *args to the end always.
# 2. assign the value of the varibales directly while calling.
```

```python
# 1. place the *args to the end always.
def test6(a , *args):
    print(a)
    return args
```

```python
test6(10,20,30,40) # 10 (20,30,40)
```

```python
# 2. assign the value of the varibale(a) directly.
def test7(*args , a):
    print(a)
    return args
```

```python
def test9(a , *args , b , c):
    print(a)
    print(args)
    print(b)
    print(c)
    return a , args , b , c
```

```python
test9(1,2,3,4,5,b=6,c=7) # default a=1 , args=(2,3,4,5) , b=6 , c-7
```

```python
def test10(a,b,c,*args):
    print(a)
    print(args)
    print(b)
    print(c)
    return a , args , b , c
```

```python
test10(1,2,3,4,5,6,7) # default a=1|b=2|c=3 , args=(4,5,6,7)
```

```python
# user can pass any data , but we should return only lists from that i/p if lists exists
def test11(*args):
    # args -> tuple
    l = []
    for i in args:
        if(type(i) == list):
            l.append(i)
    if(len(l) == 0):
        # no list data
        return -1
    else:
        return l
```

```python
test11(1,3.4,[1,2,3,4],"chandra")
```

```python
test11(1,3.4,[1,2,3,4],"chandra",[5,6,7],(1,2,3),[3,4,5])
```

```python
test11(1,2,3.4,(1,2,3))
```

```python
# create a function to take user data 
def test14(**user_data):
    print(type(user_data))
    user_details = user_data.keys()
    print(user_details)
    user_info = user_data.values()
    print(user_info)

    return user_data
```

```python
# comprehension
```

```python
l = [1,2,3,4,5]
# for this list add 2 for each element
l1 = []
for i in l :
    l1.append(i+2)
    
print(l1)
```

```python
# the same operation can be performed using comprehension of list
# list comprehension 
# var = [retun_value_logic iteration_logic condition_logic]
```

```python
l
```

```python
l2 = [i+2 for i in l]
print(l2)
```

```python
l2 = [(i*2) for i in l if i<=3]
print(l2)
```

```python
# list comprehension + lamdba expression
l3 = lambda l : [(i+2) for i in l]
print(l3(l))
```

```python
print(l3(l2))
```

```python
# create a dict comprehension where key will be a number and its vaue will be square of the number
# dict comprehension
# var = {retun_value_logic iteration_logic condition_logic}
```

```python
d1 = {i:i**2 for i in range(1,6)} # 1,2,3,4,5
print(d1)
```

```python
d2 = {}
for i in range(1,6): # 1,2,3,4,5
    d2[i] = i**2

print(d2)
```
